**Data Cleaning**

```python
df.isna().sum()
```

```python
df[df['Fuel Type'].isna()]
```

```python
df['Fuel Type'].unique()
```

```python
df['Fuel Type'].fillna('flex-fuel(unleaded/natural gas)',inplace = True)
```

```python
df['Fuel Type'].isna().sum()
```

```python
df[df['Horse Power'].isna()]
```

```python
df.groupby(by = 'Make')['Horse Power'].transform(lambda x:x.fillna(x.mean()))
```

```python
df['Horse Power'].fillna(df['Horse Power'].mean(),inplace = True) #for tesla as there is no value so mean couldnit apply in the previous step
```

```python
df.isna().sum()
```

```python
df[df['Cylinders'].isna()]
```

```python
df['Cylinders'] = df.groupby(by = 'Make')['Cylinders'].transform(lambda x:x.fillna(x.median()))
```

```python
df['Cylinders'].isna().sum()
```

```python
df[df['Number of Doors'].isna()]
```

```python
df['Number of Doors'].fillna(4,inplace = True)
```

```python
df.isna().sum()
```

```python
df[df['Market Category'].isna()]
```

```python
df['Market Category'].fillna(df['Market Category'].mode()[0],inplace = True)
```

```python
df.isna().sum()
```

```python
q1 = df['highway MPG'].quantile(0.25)
q3 = df['highway MPG'].quantile(0.75)
iqr = q3-q1
ll= max(df['highway MPG'].min(),q1-1.5*iqr)
ul = min(df['highway MPG'].max(),q3+1.5*iqr)
```

```python
df['highway MPG'] = df['highway MPG'].clip(ll,ul)
```
